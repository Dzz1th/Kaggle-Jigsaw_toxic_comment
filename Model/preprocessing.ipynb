```python
import pandas as pd 
import numpy as np
import re
from tqdm import tqdm 

TEST_PATH = '../data/test.csv'
TRAIN_PATH = '../data/train.csv'


tqdm.pandas()

train = pd.read_csv('c:/Data/data/train.csv')
test = pd.read_csv('c:/Data/data/test.csv')
print(train.shape)
print(test.shape)
```

```python
def BuildVocab(text , verbose = True):
    vocab = {}
    for sentence in tqdm(text , disable = not verbose):
        for word in sentence:
            try:
                vocab[word] += 1
            except KeyError:
                vocab[word] = 1
    return vocab
```

```python
sentences = train['comment_text'].progress_apply(lambda x: x.split()).values
vocab = BuildVocab(sentences)
print({k: vocab[k] for k in list(vocab)[:5]})
```

```python
import operator 

def CheckCoverage(vocab , embeddings_index):
    a = {}
    oov = {}
    k = 0
    i = 0
    for word in tqdm(vocab):
        try:
            a[word] = embedding_index[word]
            k += vocab[word]
        except:
            oov[word] = vocab[word]
            i += vocab[word]
            pass
    print('Found embeddings for {:.2%} of vocab'.format(len(a) / len(vocab)))
    print('Found embeddings for {:.2%} of all text'.format(k / k+i))
    sorted_x = sorted(oov.items() , key = operator.itemgetter(1))[::-1]
    
    return sorted_x
```

```python
import codecs
EMB_PATH = 'c:/Data/data/crawl-300d-2M.vec'

def GetEmbeddingCoef(word , *arr):
    return word , np.asarray(arr , dtype = 'float32')

def LoadEmbeddings(emb_path):
    embedding_index = dict(GetEmbeddingCoef(*o.strip().split(" ")) for o in tqdm(codecs.open(emb_path)))

embedding_index = LoadEmbeddings(EMB_PATH)
```

```python
sorted = ChechCoverage(vocab , embedding_index)

print(sorted[5:])
```
